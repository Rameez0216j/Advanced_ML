#### Example 2

```python
import json

json_string = '''
{
  "students": [
    {
      "name": "Millie Brown",
      "active": true,
      "rollno": 11
    },
    {
      "name": "Sadie Sink",
      "active": true,
      "rollno": 10
    }
  ]
}
'''
print(json_string)
```

```python
print("The type of object is: ", type(json_string))
```

Important NOTE : maybe json is a string and is stored in files using open operation and simply writing the json_string into it

```python
stud_obj = json.loads(json_string)
print(stud_obj)
print("The type of object is: ", type(stud_obj))
```

```python
json_obj = json.dumps(stud_obj)
print(json_obj)
print("The type of object is: ", type(json_obj))
```

```python
print(json.dumps(stud_obj,indent=4))
```
