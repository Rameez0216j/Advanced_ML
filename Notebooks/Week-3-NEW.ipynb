```python
import skimage
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io
```

### Edge Detection

```python
# Before edge detection image must be converted to grayscale
from skimage.color import rgb2gray
taj_mahal=io.imread('Tajmahal.jpg')
taj_gray=rgb2gray(taj_mahal)
io.imshow(taj_gray)
```

```python
from skimage.feature import canny
taj_edges = canny(taj_gray, sigma=3)
plt.imshow(taj_edges,cmap="binary")
```

#### Detecting edges

```python
from skimage.feature import canny
taj_edges = canny(taj_gray, sigma=3)
fig = plt.figure(figsize = (16,9))
ax1 = fig.add_subplot(1,2, 1)
ax1.imshow(taj_edges, cmap='binary')
ax2 = fig.add_subplot(1,2, 2)
ax2.imshow(taj_gray, cmap='binary')
```
